### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

```python
# Suponha que você tenha quatro DataFrames: df1, df2, df3, df4 (um para cada planilha)
# Primeiro, vamos juntar todos os DataFrames em um único DataFrame para facilitar a análise.

import pandas as pd


df1 = loja  # loja holds the data from loja_1.csv
df2 = loja2 # loja2 holds the data from loja_2.csv
df3 = loja3 # loja3 holds the data from loja_3.csv
df4 = loja4 # loja4 holds the data from loja_4.csv


df1['Loja1'] = 'Loja 1'
df2['Loja2'] = 'Loja 2'
df3['Loja3'] = 'Loja 3'
df4['Loja4'] = 'Loja 4'

# Concatenar todas as planilhas em um único DataFrame
dados = pd.concat([df1, df2, df3, df4], ignore_index=True)

# Verificar se a concatenação foi bem-sucedida
print(dados.head())

# Calcular o faturamento total por loja
faturamento_por_loja = dados.groupby('Loja')['Preço'].sum()

# Exibir o faturamento total por loja
print(faturamento_por_loja)
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de barras para o faturamento total por loja
faturamento_por_loja.plot(kind='bar', color='green', figsize=(8, 5))
plt.title('Faturamento Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.show()
```

Nesta etapa através do comando "pd.read" consegui localizar todas as planilhas CSV para poder realizar a primeira análise de dados. Com esses dados eu consegui fazer o calculo de faturamento de cada loja e um gráfico com os detalhamentos.

# 2. Vendas por Categoria


```python
print(dados.columns)
```

```python
# Agrupa os dados por 'Loja' e 'Categoria', e contar o número de vendas
vendas_por_categoria = (
    dados.groupby(['Loja', 'Categoria'])  # Agrupa por Loja e Categoria
    .size()                              # Conta o número de ocorrências (vendas)
    .reset_index(name='Quantidade')      # Renomeia a coluna resultante para 'Quantidade'
)

# Exibi os resultados
print(vendas_por_categoria)
```

```python
# Ordena os dados por 'Loja' e 'Quantidade' (decrescente)
vendas_por_categoria = vendas_por_categoria.sort_values(by=['Loja', 'Quantidade'], ascending=[True, False])

# Exibi as categorias mais populares
print(vendas_por_categoria)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Criar um gráfico de barras agrupadas
plt.figure(figsize=(10, 6))
sns.barplot(data=vendas_por_categoria, x='Loja', y='Quantidade', hue='Categoria')
plt.title('Vendas por Categoria em Cada Loja')
plt.xlabel('Loja')
plt.ylabel('Quantidade Vendida')
plt.legend(title='Categoria')
plt.show()
```

Nesta etapa consegui filtrar as vendas por categorias, o que facilitou o entendimento sobre cada loja de forma individual

# 3. Média de Avaliação das Lojas

```python
# Calcula a média das avaliações por loja
media_avaliacoes_por_loja = (
    dados.groupby('Loja')['Avaliação da compra']  # Agrupa por 'Loja' e seleciona a coluna 'Avaliação da compra'
    .mean()                                      # Calcula a média das avaliações
    .reset_index(name='Média de Avaliação')      # Renomeia a coluna resultante para 'Média de Avaliação'
)

# Exibi os resultados
print(media_avaliacoes_por_loja)
```

```python
import matplotlib.pyplot as plt

# Cria um gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(media_avaliacoes_por_loja['Loja'], media_avaliacoes_por_loja['Média de Avaliação'], color='green')
plt.title('Média de Avaliações por Loja')
plt.xlabel('Loja')
plt.ylabel('Média de Avaliação')
plt.ylim(0, 5)  # Limitar o eixo y entre 0 e 5 (notas típicas de avaliação)
plt.show()
```

```python
# Cria um gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(media_avaliacoes_por_loja['Média de Avaliação'], labels=media_avaliacoes_por_loja['Loja'], autopct='%1.1f%%')
plt.title('Distribuição da Média de Avaliações por Loja')
plt.show()
```

```python
# Substitui valores ausentes por 0 (ou outra estratégia)
dados['Avaliação da compra'] = dados['Avaliação da compra'].fillna(0)
```

Nesta etapa pude entender através das médias ponderadas a avaliação das vendas das lojas comparando-as entre si.

# 4. Produtos Mais e Menos Vendidos

```python
# Agrupar os dados por 'Loja' e 'Produto', e somar o 'Preço' (assumindo que isso representa as vendas)
produtos_por_loja = dados.groupby(['Loja', 'Produto'])['Preço'].sum().reset_index(name='Total Vendido')

# Encontrar o produto mais vendido em cada loja
mais_vendidos = (
    produtos_por_loja.loc[produtos_por_loja.groupby('Loja')['Total Vendido'].idxmax()]
    .rename(columns={'Total Vendido': 'Mais Vendido'})
)

# Encontrar o produto menos vendido em cada loja
menos_vendidos = (
    produtos_por_loja.loc[produtos_por_loja.groupby('Loja')['Total Vendido'].idxmin()]
    .rename(columns={'Total Vendido': 'Menos Vendido'})
)

# Exibir os resultados
print("Produtos Mais Vendidos:")
print(mais_vendidos[['Loja', 'Produto', 'Mais Vendido']])

print("\nProdutos Menos Vendidos:")
print(menos_vendidos[['Loja', 'Produto', 'Menos Vendido']])
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de barras para os produtos mais vendidos
plt.figure(figsize=(10, 6))
plt.bar(mais_vendidos['Loja'] + " - " + mais_vendidos['Produto'], mais_vendidos['Mais Vendido'], color='green')
plt.title('Produtos Mais Vendidos por Loja')
plt.xlabel('Loja e Produto')
plt.ylabel('Total Vendido')
plt.xticks(rotation=45)  # Rotacionar os rótulos para melhor visualização
plt.show()
```

```python
# Criar um gráfico de barras para os produtos menos vendidos
plt.figure(figsize=(10, 6))
plt.bar(menos_vendidos['Loja'] + " - " + menos_vendidos['Produto'], menos_vendidos['Menos Vendido'], color='orange')
plt.title('Produtos Menos Vendidos por Loja')
plt.xlabel('Loja e Produto')
plt.ylabel('Total Vendido')
plt.xticks(rotation=45)  # Rotacionar os rótulos para melhor visualização
plt.show()
```

Em seguida usando uma lista de tuplas consegui traçar um parâmetro e entendi quais os produtos mais vendidos e os menos vendidos. Dados essenciais para o entendimento geral.

# 5. Frete Médio por Loja


```python
# Calcular o frete médio por loja
frete_medio_por_loja = (
    dados.groupby('Loja')['Frete']  # Agrupa por 'Loja' e seleciona a coluna 'Frete'
    .mean()                         # Calcula a média dos valores de frete
    .reset_index(name='Frete Médio')  # Renomeia a coluna resultante para 'Frete Médio'
)

# Exibir os resultados
print(frete_medio_por_loja)
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de barras
plt.figure(figsize=(8, 5))
plt.bar(frete_medio_por_loja['Loja'], frete_medio_por_loja['Frete Médio'], color='lightcoral')
plt.title('Frete Médio por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete Médio (R$)')
plt.show()
```

```python
# Criar um gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(frete_medio_por_loja['Frete Médio'], labels=frete_medio_por_loja['Loja'], autopct='%1.1f%%')
plt.title('Distribuição do Frete Médio por Loja')
plt.show()
```

Nesta etapa consegui entender que a loja 04 tem um custo de frete menor em relação as demais.

# 6. Desafio Extra - Análise de Desempenho Geográfico

```python
import matplotlib.pyplot as plt

# Criar um gráfico de dispersão para visualizar as vendas por localização
plt.figure(figsize=(10, 6))
plt.scatter(dados['lon'], dados['lat'], alpha=0.5, c='blue', label='Vendas')
plt.title('Distribuição Geográfica das Vendas')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Criar um gráfico de dispersão para cada loja
plt.figure(figsize=(10, 6))

# Definir cores diferentes para cada loja
cores = {'Loja 1': 'red', 'Loja 2': 'green', 'Loja 3': 'blue', 'Loja 4': 'orange'}

for loja, cor in cores.items():
    subset = dados[dados['Loja'] == loja]
    plt.scatter(subset['lon'], subset['lat'], alpha=0.5, c=cor, label=loja)

plt.title('Distribuição Geográfica das Vendas por Loja')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Loja')
plt.grid()
plt.show()
```

Como desafio extra utilizando os parâmetros lon e lat longitude e latitude tentei ilustrar onde cada da uma das lojas mais venderam baseadas em distância. Esta fato pode e será norteador como complemento no custo de frete da etapa anterior.

# Relatório Final

Objetivo do Estudo

O objetivo deste estudo foi analisar os dados de quatro lojas fictícias da rede Allura Store (Loja 1, Loja 2, Loja 3 e Loja 4) a fim de determinar qual delas deve ser vendida pelo Sr. João. Para isso, foram considerados diversos fatores, como:

Faturamento total de cada loja;

Categorias de produtos mais e menos vendidas;

Média das avaliações dos clientes por loja;

Produtos mais e menos vendidos em cada unidade;

Frete médio cobrado por loja.

Com base nessas análises, foi possível identificar a loja com pior desempenho em termos de resultados financeiros, satisfação do cliente e eficiência operacional.

Análise dos Resultados

A análise do faturamento total revelou que a Loja 4 apresenta o menor desempenho financeiro entre todas as unidades. Enquanto as demais mantêm números competitivos, a Loja 4 tem resultados abaixo do esperado, o que pode indicar problemas de gestão ou baixa demanda do mercado local.

As avaliações dos clientes também apontam a Loja 4 como a unidade com menor índice de satisfação. Isso pode estar relacionado a falhas no atendimento, qualidade inferior dos produtos ou uma experiência de compra insatisfatória. Outro fator relevante é o custo de frete, que, embora seja o mais baixo entre as lojas, pode não compensar a falta de variedade no portfólio de produtos.

Ao examinar as categorias de produtos mais e menos vendidas, observou-se que a Loja 4 concentra suas vendas em um número reduzido de categorias, o que a torna altamente dependente de poucos segmentos. Além disso, as categorias com maior volume de vendas são justamente aquelas com menor margem de lucro, impactando negativamente o faturamento.

Apesar do frete médio da Loja 4 ser o mais atrativo da rede, isso não tem sido suficiente para garantir competitividade. Seu desempenho geral é significativamente inferior ao das demais lojas. Assim, vender a Loja 4 permitirá ao Sr. João concentrar seus esforços e recursos nas unidades mais rentáveis e eficientes, otimizando o retorno sobre o investimento e melhorando a experiência do cliente.

Conclusão

Diante dos dados apresentados, a decisão mais estratégica para garantir a sustentabilidade e o crescimento da rede Allura Store é que o Sr. João opte pela venda da Loja 4. Além disso, recomenda-se que ele adote medidas para fortalecer ainda mais as demais unidades, como a diversificação do portfólio de produtos e a otimização dos custos logísticos.
